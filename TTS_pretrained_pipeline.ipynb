```python

def measure_latency_and_wer(texts, tts_model, vocoder_model,output_dir):
    """
    Synthesizes speech from text and measures the performance
  
    Args:
    texts: A list containing texts to be synthesized into speech
    tts_model: the path of the tts spectrogram model
    vocoder_model: the path of the tts vocoder model
    ouput_dir: the directory to store the results

    Returns:
    df: A dataframe containing the evaluated metrics and model name

    """
  
  
    os.environ["CUDA_VISIBLE_DEVICES"]="0"
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
    
    model = whisper.load_model("large-v2", device = DEVICE) # Select whisper model size ['tiny.en', 'tiny', 'base.en', 'base', 'small.en', 'small', 'medium.en', 'medium', 'large-v1', 'large-v2', 'large']
    tts_name=os.path.basename(tts_model)
    voc_name=os.path.basename(vocoder_model)
    

    processing_times = [] # Create an empty list to store the latency scores
    WER = []   # Create an empty list to store the WER scores
    mos_scores = [] # Create an empty list to store the MOS scores

    for i, text in enumerate(texts):
        # Set the output path for this iteration
        output_path = f"{output_dir}/{tts_name}_{voc_name}_output{i+1}.wav"

        # Run the TTS model and capture the output
        output_bytes = subprocess.check_output(['tts', '--model_name', tts_model, '--vocoder_name', vocoder_model, '--text', text, '--out_path', output_path])

        # Decode the byte string into a regular string
        output_str = output_bytes.decode('utf-8')

        # Extract the processing time from the output string
        output_lines = output_str.strip().split('\n')
        processing_time_line = output_lines[-3]
        processing_time = float(processing_time_line.split(': ')[1])

        # Append the processing time to the list
        processing_times.append(processing_time)
        # Measuring the WER
        audio = whisper.load_audio(output_path)

        # Transcribe audio
        result = model.transcribe(audio)

        # Print transcript
        print(f'transcribed text of {tts_name}_{voc_name}_output{i+1}:{result["text"]}')

        ref_text = text

        # Compute WER
        ref_text=jiwer.ToLowerCase()(ref_text)
        transcribed_text=jiwer.ToLowerCase()(result["text"])
        error = jiwer.wer(ref_text, transcribed_text)
        # Append the WER to the list
        WER.append(error*100)


    %cd /content/MOSNet
    # Call the custom_test.py script and capture its output
    !source /content/mosnet/bin/activate; python ./custom_test.py --rootdir "{output_dir}"
    mos_output_file=f'{output_dir}/MOSnet_result_raw.txt'
    # Open the file for reading
    with open(mos_output_file, 'r') as f:
        # Loop through each line in the file
        for line in f:
            # Split the line into its path and MOS score using the whitespace as a delimiter
            path, score = line.strip().split(' ')
            # Append the score to the MOS scores list
            mos_scores.append(float(score))
          
    %cd /content
    # Calculate the mean and standard deviation of the MOS scores using numpy
    mean_mos = np.mean(mos_scores)
    std_dev_mos = np.std(mos_scores)
    # Calculate the mean and standard deviation of the processing times
    mean_latency = np.mean(processing_times)
    std_deviation_latency = np.std(processing_times)
    # caculate the mean and standard deviation of the MOS scores
    mean_WER = np.mean(WER)
    std_WER = np.std(WER)

    # Add the model name, mean and standard deviation of latency and WER to a pandas dataframe
    df = pd.DataFrame({
        'Model': [f'{tts_name} with {voc_name}'],
        'Mean Latency (s)': [mean_latency],
        'Std. Deviation Latency (s)': [std_deviation_latency],
        'Mean WER (%)': [mean_WER],
        'Std. Deviation WER (%)': [std_WER],
        'Mean MOS': [mean_mos],
        'Standard Deviation of MOS': [std_dev_mos]
    })
    os.makedirs('/content/eval_output', exist_ok=True)
    eval_dir='/content/eval_output'
    # Write the dataframe to a CSV file
    output_file = f"{tts_name}_{voc_name}_output.csv"
    output_path = os.path.join(eval_dir, output_file)
    df.to_csv(output_path, index=False)
    # Print the results
    print(f"Mean latency: {mean_latency:.2f}")
    print(f"Standard deviation of latency: {std_deviation_latency:.2f}")
    print(f"Mean WER: {mean_WER}")
    print(f"Standard deviation of WER: {std_WER}")
    print('Mean MOS:', mean_mos)
    print('Standard Deviation of MOS:', std_dev_mos)
    
    # Return the dataframe and the list of output paths
    return df
```
